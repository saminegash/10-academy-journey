#### Normalize each engagement metric and run a k-means (k=3) to classify customers in three groups of engagement.

```python
from sklearn import preprocessing
from sklearn.cluster import KMeans
import scipy.spatial.distance as sdist
```

```python
# Creating unnormalized Data
df_unnormalized = challenge_data.groupby('MSISDN/Number').agg(dict( {'Bearer Id': 'count', 
                                                                       'Dur. (ms)': 'sum',
                                                                       'total_data_trans':'sum'})
                                                               )
```

```python
unnormalized_data = df_unnormalized.rename(columns={'Bearer Id':'session_frequency', 'Dur. (ms)': "session_duration", 'total_data_trans':"Total Traffic" })
unnormalized_data.head(10)
```

```python
# Create df, where df the 3 column's values as floats
df = unnormalized_data[['session_frequency','session_duration', 'Total Traffic']].values.astype(float)


# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
df_scaled = min_max_scaler.fit_transform(df)

# Run the normalizer on the dataframe
normalized_data = pd.DataFrame(df_scaled)
```

```python
normalized_data.head(10)
```

```python
norm_data = ((unnormalized_data - unnormalized_data.min())/(unnormalized_data.max() - unnormalized_data.min()))*1
    
norm_data.head()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++').fit(norm_data)
unnormalized_data['cluster'] = kmeans.labels_
```

### Finding the centroid
