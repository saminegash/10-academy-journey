### Variable transformations – segment the users into top five decile classes
* based on the total duration for all sessions and compute the total data
* (DL+UL) per decile class. - (jupyter notebook + slide )

```python
duration_new = challenge_data.groupby(['Bearer Id']).sum() # duration to use for non graphical analysis
```

```python
duration_new
```

```python
duration_new['Decile_class'] = pd.qcut(duration_new['Dur. (ms)'], 5, 
                            labels = False) 
```

```python
df = duration_new[['Dur. (ms)','Decile_class','total_data_trans']]
```

```python
df['total_data_trans_in_Tera_byte'] = df['total_data_trans']/(10**12)
df.head()
```

```python
df.groupby(['Decile_class'])['total_data_trans_in_Tera_byte'].sum()
```

```python
fig, ax = plt.subplots(ncols=1)
df.groupby(['Decile_class'])['total_data_trans_in_Tera_byte'].sum().plot.bar()
ax.set_title('Bar plot of Decile Class vs total data transfer in Tera bytes')
ax.set_xlabel('Top 5 Decile Class')
ax.set_ylabel('Total Data Transfered in Tera Bytes')
fig.savefig('Data Segmentation Decile.png')
plt.show()
```
